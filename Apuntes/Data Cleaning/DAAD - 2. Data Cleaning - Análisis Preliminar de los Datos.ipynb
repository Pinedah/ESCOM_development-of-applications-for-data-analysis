```python
data = data.drop(["sex"], axis = 1) #eliminamos columna sex
```

```python
data = pd.concat([data, dummy_sex], axis = 1) #insertamos dummies
```

```python
def createDummies(df, var_name): #funcion para crear variables dummies
    dummy = pd.get_dummies(df[var_name], prefix=var_name)
    df = df.drop(var_name, axis = 1)
    df = pd.concat([df, dummy ], axis = 1)
    return df
```
